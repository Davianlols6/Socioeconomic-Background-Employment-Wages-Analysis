### Does a person's gender affect the wages they get?

```python
# Copy the dataset so as to not alter the original dataset
meanEarningsBySex_copy = meanEarningsBySex.copy()

# Pivoting the table
meanEarningsBySex_copy = meanEarningsBySex_copy.pivot(index='quarter', columns='sex', values='average_monthly_earnings')

import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))

# Plot the dataset
plt.plot(meanEarningsBySex_copy['male'], marker='o', label='Male')
plt.plot(meanEarningsBySex_copy['female'], marker='o', label='Female')

plt.legend()
plt.title("Mean Earnings By Sex")

plt.xticks(rotation=90)
plt.xlabel("Quater")
plt.ylabel("Earnings ($)")
plt.show()
```

We can see that male workers earn a higher wage on average compared to their female counterparts. We can also see that the average earnings for both genders are on an increasing trend. We can also see that the mean salary usually rise and fall around Quarter 4 and Quarter 1.

One possible explanation for this fluctuation is that bonuses are usually given out at the end or at the start of the year, therefore and increase in the mean salary during these periods of the year is expected.

```python
meanEarningsBySex_copy['gender_difference'] = meanEarningsBySex_copy['male'] - meanEarningsBySex_copy['female']

import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
# Plot the dataset
plt.plot(meanEarningsBySex_copy['gender_difference'], marker='o', label='Difference')
plt.legend()
plt.title("Difference In Earnings Between Males And Females")

plt.xticks(rotation=90, fontsize=8)
plt.xlabel("Quater")
plt.ylabel("Earnings ($)")
plt.show()
```

We can see that the difference in the wage gap between males and females was initially on an increasing trend from 1988 to about 2014 but stopped and remained stable there from 2014 till 2023.

The difference in pay between both genders in recent years are between $1.2k to $1.6k

### Wage breakdown by industry

```python
# Copy the dataset so as to not alter the original dataset
medianWageOfOccupationsbySexinAllIndustries_copy = medianWageOfOccupationsbySexinAllIndustries.copy()

# Get rid of occupation which only has data for 1 gender
medianWageOfOccupationsbySexinAllIndustries_copy = medianWageOfOccupationsbySexinAllIndustries_copy[medianWageOfOccupationsbySexinAllIndustries_copy['occ_desc'].duplicated(keep=False)]

# Pivoting the table
medianWageOfOccupationsbySexinAllIndustries_copy = medianWageOfOccupationsbySexinAllIndustries_copy.pivot_table(index='occ_desc', columns='sex2', values='mthly_basic_wage_50_pctile', aggfunc='first')

# Calculating the wage gap per industry
medianWageOfOccupationsbySexinAllIndustries_copy['wage_difference'] = abs(medianWageOfOccupationsbySexinAllIndustries_copy['males'] - medianWageOfOccupationsbySexinAllIndustries_copy['females'])

# Sort rows by wage_difference
medianWageOfOccupationsbySexinAllIndustries_copy.sort_values('wage_difference', inplace=True, ascending=False)

# Getting the top 10 industries with the biggest and smallest pay gap
medianWageOfOccupationsbySexinAllIndustries_copy_top10 = medianWageOfOccupationsbySexinAllIndustries_copy.head(10)
medianWageOfOccupationsbySexinAllIndustries_copy_bottom10 = medianWageOfOccupationsbySexinAllIndustries_copy.tail(10)[::-1]

# Plot the dataset
import matplotlib.pyplot as plt

x = np.arange(len(medianWageOfOccupationsbySexinAllIndustries_copy_top10))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, (ax, ax2) = plt.subplots(1,2,layout='constrained')
fig.set_size_inches(16, 9)

for attribute, measurement in medianWageOfOccupationsbySexinAllIndustries_copy_top10[['males', 'females']].items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Salary ($)')
ax.set_xlabel('Industry')
ax.set_title('Top 10 Professions With The Biggest Gender Pay Gap')
ax.set_xticks(x + width, medianWageOfOccupationsbySexinAllIndustries_copy_top10.index, rotation=90)
ax.legend(loc='upper left', ncols=3)

multiplier = 0

for attribute, measurement in medianWageOfOccupationsbySexinAllIndustries_copy_bottom10[['males', 'females']].items():
    offset = width * multiplier
    rects = ax2.bar(x + offset, measurement, width, label=attribute)
    ax2.bar_label(rects, )
    multiplier += 1

ax2.set_ylabel('Salary ($)')
ax2.set_xlabel('Industry')
ax2.set_title('Top 10 Professions With The Smallest Gender Pay Gap')
ax2.set_xticks(x + width, medianWageOfOccupationsbySexinAllIndustries_copy_bottom10.index, rotation=90)
ax2.legend(loc='upper left', ncols=3)
plt.show()
```

We can see that the industry with the highest gender pay gap are mostly finance related and are usually high paying jobs like sales professionals, fund/portfolio manager and the likes. Industries with higher gender pay gap are usually considered high-level jobs and have higher pay. On the other end, industries with the smallest gender pay gap are those usually considered to be low-level like table-top cleaner, security supervisor and dishwasher.

### Unemployment By Highest Qualification Attained

```python
# Copy the dataset so as to not alter the original dataset
unemployedResidentsByHighestQualificationAttained_copy = unemployedResidentsByHighestQualificationAttained.copy()

# Group by 'year', 'sex', and 'highest_qualification' and sum the 'labour_force'
result = unemployedResidentsByHighestQualificationAttained_copy.groupby(['year', 'highest_qualification'], as_index=False)['unemployed'].sum()

# Getting the x-values for each qualifications
highest_qualification_degree = result[result['highest_qualification'] == 'degree']
highest_qualification_diploma = result[result['highest_qualification'] == 'diploma and professional qualification']
highest_qualification_lower_secondary = result[result['highest_qualification'] == 'lower secondary']
highest_qualification_post_sec = result[result['highest_qualification'] == 'post-secondary (non-tertiary)']
highest_qualification_primary_and_below = result[result['highest_qualification'] == 'primary and below']
highest_qualification_secondary = result[result['highest_qualification'] == 'secondary']

import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))

# Plot the dataset
plt.plot(highest_qualification_degree['year'], highest_qualification_degree['unemployed'], marker='o', label='Degree')
plt.plot(highest_qualification_degree['year'], highest_qualification_diploma['unemployed'], marker='o', label='Diploma')
plt.plot(highest_qualification_degree['year'], highest_qualification_lower_secondary['unemployed'], marker='o', label='Lower Secondary')
plt.plot(highest_qualification_degree['year'], highest_qualification_post_sec['unemployed'], marker='o', label='Post Secondary')
plt.plot(highest_qualification_degree['year'], highest_qualification_primary_and_below['unemployed'], marker='o', label='Primary And Below')
plt.plot(highest_qualification_degree['year'], highest_qualification_secondary['unemployed'], marker='o', label='Secondary')

plt.legend()
plt.title("Unemployment By Highest Qualification Attained")

plt.xlabel("Years")
plt.ylabel("Unemployed")
plt.ticklabel_format(style='plain') 
plt.show()
```

We can see that the unemployment rate for degree and diploma holders have been on an increasing trend since 2010 while other qualifications have remained stable or even decreased. This is unexpected as we would expect that degree or diploma holders would have a lower unemployment rate due to their high qualifications. One possible reason for this would be that there is an oversaturation of degree and diploma holders.

### How does age affect employment opportunities?

```python
# Copy the dataset so as to not alter the original dataset
labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy = labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex.copy()

# Get only useful columns
labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy = labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy[labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex.year == 2022][['age', 'labour_force', 'outside_labour_force']]

# Group by the dataset by age
labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy = labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy.groupby(['age'], as_index=False).sum()

# Melt the table
labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy = pd.melt(labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy, id_vars=['age'], var_name='category', value_name='value')

import matplotlib.pyplot as plt
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8), (ax9, ax10, ax11, ax12)) = plt.subplots(3, 4, figsize=(16,9))

axList = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10, ax11, ax12]

count = 0
for i in labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy['age'].unique():
    ax = axList[count]
    x = labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy[labourForceAged15YearsandOverbyLabourForceStatusAgeAndSex_copy['age'] == i]
    ax.pie(x['value'], labels=x['category'], autopct='%1.1f%%')
    ax.set_title(f"Labour Force Participation\nBy Age {i}")
    count += 1
```

As we can see from the pie charts above, only a minority, 15.9% of youths aged 15 to 19 are working. This is expected as the majority of youth in that age range are currently still studying full time. The labour force aged 25-29 shows the completely opposite result with the minority, 10.7% of that age group outside of the work force. From ages 25-29 onwards, the proportions of labour force participations remains roughly the same until the ages 50-54 where the proportion of those outside the work force starts to grow. This is expected as from that age group onwards, people would start to retire. The biggest jump in outside work force proportion is from age 60-64 to age 65-69 from 32.5& to 50.6%. This is expected as the retiring age in Singapore is 63 years old. Only 21.8% of those 70 and over are still working.

### How does unemployment rate affect wages? 

#### How does unemployment rate affect total wage changes?

```python
# Copy the dataset so as to not alter the original dataset
overallUnemploymentRate_copy = overallUnemploymentRate.copy()
totalWageChange_copy = totalWageChange.copy()

# Drop unnecessary columns
overallUnemploymentRate_copy.drop('residential_status', inplace=True, axis=1)

# Merge with total wage change dataset
overallUnemploymentRate_copy = overallUnemploymentRate_copy.merge(totalWageChange_copy, on='year', how='inner')

# Get the OLS regression result
unemploymentRateByTWC = smf.ols(formula='unemployment_rate ~ twc', data=overallUnemploymentRate_copy).fit()
```

```python
print(unemploymentRateByTWC.summary())
```

```python
x=overallUnemploymentRate_copy['unemployment_rate']
y=overallUnemploymentRate_copy['twc']

import matplotlib.pyplot as plt
fig1, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16,9))

# matplotlib scatterplot
ax1.scatter(y=y, x=x)
# best fit line Y = C + M * X
m,c = np.polyfit(x, y, deg=1)
ax1.plot(x, c+m*x, 'r-')
ax1.set_xlabel('Unemployment Rate')
ax1.set_ylabel('Total Wage Change')

ax1.set_title("Total Wage Changes vs Unemployment Rate")

linearModEq = "Y = {:.5} {:.5}X".format(unemploymentRateByTWC.params[0],unemploymentRateByTWC.params[1])
ax1.annotate(linearModEq, xy=(2.3, 4.109937740883488), xytext=(2.5,5), size=12, arrowprops=dict(facecolor='green', shrink=0.05))

sns.histplot(overallUnemploymentRate_copy['twc'], kde=True, ax=ax2)
sm.qqplot(overallUnemploymentRate_copy['twc'], line='s', ax=ax3)

plt.show()
```

Total wage change includes bonuses along with the basic pay. 
With a squared value of 0.465, we can see that there is a moderate negative relationship between the total wage change and unemployment rate. This goes to show that as the unemployment rate rises, wages rises at a slower pace and lesser bonuses is given out, which is to be expected. 
For every one percent increase in unemployment rate, the total wage change decreases by 0.22%. 
The distribution is not close to normal as can be seen in the histogram and Q-Q plot. Which can lead to the model not fitting well to the dataset.
